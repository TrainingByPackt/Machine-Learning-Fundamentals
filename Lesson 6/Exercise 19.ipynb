```python
import pandas as pd
```

```python
data = pd.read_csv("../datasets/fertility_Diagnosis.csv", header=None)
```

```python
X = data.iloc[:,:9]
Y = data.iloc[:,9]
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(max_iter = 1200)
model.fit(X,Y)
```

```python
import pickle
import os
```

```python
path = os.getcwd()+'/model_exercise.pkl'
file = open(path, 'wb')
pickle.dump(model, file)
```
